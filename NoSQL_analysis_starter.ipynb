### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene' : 20}
fields = {}
# Use count_documents to display the number of documents in the result
clean_e = establishments.count_documents(query)

print(f'There are {clean_e} establishments with a hygiene score of 20')
# Display the first document in the results using pprint
clean_es = establishments.find(query, fields)
pprint(clean_es[0])
```

```python
# Convert the result to a Pandas DataFrame
clean_es_list = list(clean_es)
clean_df = pd.DataFrame(clean_es_list)
clean_df
# Display the number of rows in the DataFrame
row_count = len(clean_df)
print(row_count)
```

```python
# Display the first 10 rows of the DataFrame
clean_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
#First find the title London has under the 'LocalAuthorityName' Column
column_name = "LocalAuthorityName"

# Find all the unique strings for the specified column
unique_strings = establishments.distinct(column_name)

# Print the unique strings
print(unique_strings)
```

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {
    'LocalAuthorityName' : 'City of London Corporation',
    'RatingValue' : {'$gte': 4}
}
fields = {}
# Use count_documents to display the number of documents in the result
print(f'There are {establishments.count_documents(query)} establishments that have London as the Local Authority and a Rating of 4 or higher.')
# Display the first document in the results using pprint
docs = establishments.find(query, fields)
pprint(docs[0])
```

```python
# Convert the result to a Pandas DataFrame
cleaned_docs = list(docs)
df = pd.DataFrame(cleaned_docs)
```

```python
# Display the number of rows in the DataFrame
row_count = len(df)
print(row_count)
# Display the first 10 rows of the DataFrame
df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

query = {
    'RatingValue': 5,
    'geocode.latitude': {
        '$gte': latitude - degree_search,
        '$lte': latitude + degree_search
    },
    'geocode.longitude': {
        '$gte': longitude - degree_search,
        '$lte': longitude + degree_search
    }
}

# Print the results
results = establishments.find(query).sort('scores.Hygiene' , 1).limit(5)
result_list = list(results)

for result in result_list:
    pprint(result)
```

```python
# Convert result to Pandas DataFrame
df = pd.DataFrame(result_list)
df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene' : 0}}
# 2. Groups the matches by Local Authority
group_query = {'$group' : {'_id' : '$LocalAuthorityName'}}
# 3. Sorts the matches from highest to lowest
sort_values = {'$sort' : {'RatingValue': -1}}
# Print the number of documents in the result
pipeline = [match_query, group_query, sort_values]
results = list(establishments.aggregate(pipeline))
print(len(results))
# Print the first 10 results
for result in results[:10]:
    print(result)
```

```python
# Convert the result to a Pandas DataFrame
df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
len(df)
# Display the first 10 rows of the DataFrame
df.head(10)
```
